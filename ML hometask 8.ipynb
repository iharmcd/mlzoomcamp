```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import SGD

# Define the model
model = Sequential([
    # Convolutional Layer
    Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)),
    
    # Max Pooling Layer
    MaxPooling2D(pool_size=(2, 2)),
    
    # Flattening Layer
    Flatten(),
    
    # Dense Layer with 64 neurons
    Dense(64, activation='relu'),
    
    # Output Layer with 1 neuron (for binary classification)
    Dense(1, activation='sigmoid')  # 'sigmoid' for binary classification
])

# Compile the model
optimizer = SGD(learning_rate=0.002, momentum=0.8)  # learning_rate instead of lr in newer Keras versions
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',  # Appropriate loss function for binary classification
              metrics=['accuracy'])

# Print the model summary to verify the structure
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_dir = 'data/train'
train_datagen = ImageDataGenerator(rescale=1./255)


train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(200, 200),
    batch_size=20,
    class_mode='binary',
    shuffle=True
)
```

```python
test_dir = 'data/test'
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

test_generator = test_datagen.flow_from_directory(
   test_dir,  # Replace with your actual path
    target_size=(200, 200),
    batch_size=20,
    class_mode='binary',
    shuffle=True
)


# Train the model for 10 epochs
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)

# Calculate the median training accuracy
train_acc = history.history['accuracy']
median_train_acc = np.median(train_acc)

print(f"Median Training Accuracy: {median_train_acc:.2f}")
```

```python
train_loss = history.history['loss']
std_dev_train_loss = np.std(train_loss)

print(f"Standard Deviation of Training Loss: {std_dev_train_loss:.4f}")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create a data generator with augmentation for training data
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,             # Normalize pixel values
    rotation_range=50,             # Rotate images up to 50 degrees
    width_shift_range=0.1,         # Shift width up to 10%
    height_shift_range=0.1,        # Shift height up to 10%
    zoom_range=0.1,                # Zoom up to 10%
    horizontal_flip=True,          # Randomly flip images horizontally
    fill_mode='nearest'            # Fill missing pixels with the nearest value
)

# Flow from the training directory
train_generator = train_datagen.flow_from_directory(
    train_dir,     # Replace with your training directory path
    target_size=(200, 200),        # Resize images to match the input shape
    batch_size=20,
    class_mode='binary',
    shuffle=True
)
```

```python
# Train for 10 more epochs using augmented data
history_aug = model.fit(
    train_generator,       # Augmented training data generator
    epochs=10,             # Additional epochs
    validation_data=test_generator
)

# Combine validation loss from previous and current training
val_loss_combined = history.history['val_loss'] + history_aug.history['val_loss']

# Calculate the mean of the combined validation loss
mean_val_loss = np.mean(val_loss_combined)

print(f"Mean of Test (Validation) Loss: {mean_val_loss:.4f}")
```
